#Que es Python


Numpy es una biblioteca de python creada en 2006, la cual significa numerical python, su objetivo principal es optimizar el trabajo con grandes volumenes de datos númericos, permitiendo a los cientificos y desarrolladores manipular y analizar datos de manera más rápida y eficiente.

NumPy introduce una estrucura de datos llamados ndarray, que es una array multidimensional optimizado para operaciones númericas, estos arrays son similares a las listas de python, pero con la restriccion de que todos los elementos deben ser el mismo tipo de dato, lo que permite un almacenamiento más eficiente y operaciones más rápidas, especialmente cuando se trabaja con grandes bolumenes de datos

**Beneficios de NumPy**

Arreglos multidimensionaes: Numpy permite la creación de arrays de una, dos o más dimesniones, facilitado el manejo de datos complejos como matrices y tensores

operaciones matematicas rapidas: las operaciones sobre ndarray están altamente optimizadas permitiendo realizar cálculos matemáticos y estadísticos de manera rápida

Compatibilidad con otras librerias: NumPy es la base sobre la cual se construyen muchas otras librerías populares en el ecosistema de Python para ciencia de datos, como Pandas, SciPy y scikit-learn



Arreglos: Son listas mas rapidas

np.arange: Se usa para crear un array desde cero, como si fueras generando una lista de números pero en formato NumPy.

#Numeros aleatorios


```python
# Utilizar rg.random para generar un ndarray de 1000 números aleatorios entre
# el 0 y 1

rg = np.random.default_rng(2) #Requisito de aleatoridad para poder tener los mismos números se llama semilla
```

```python
#Por haber usado el número dos todos los que lo generen tendrian el mismo número
aleatorio = rg.random(1000)
print(aleatorio)
```

```python
#Graficar los datos en 100 baldes
plt.hist(aleatorio,bins = 100)
plt.show()
```

```python
#Usar normal para generar valores aleatorios que sigan una distribución normal

normal = rg.normal(10,5,100000)
normal
```

```python
plt.hist(normal,bins = 100)
plt.show()
```

```python
#Solo numeros aleatorios enteros, no es muy utilizado
enteros = rg.integers(20, size = 2000)
print(enteros)
```

```python
#Grafico

plt.hist(enteros)
plt.show()
```

```python
# 10 Numeros aleatorios sin repetir entre el 0 y el 25

rg.choice(26,  size=10, replace = False) #26 es el total de numeros
                                       #10 los números que quiero agarrar
                                       #replace, es para que no se repitan los números
```

#Seleccionar datos en un arreglo

```python
enteros = rg.integers(20, size = (10))
enteros
```

```python
#Seleccionar los primeros 6 valores enteros del ndarray
enteros[0:6] #0 es el indice por donde comienza python
             #6 hasta donde quiero llegar, en realidad sería del 0,1,2,3,4,5
```

```python
#Seleccionar los valores en la posición 0,2 y 4 del ndarray "enteros"
enteros[0:6:2] #seleccionar el valor 0 (0)
               #Hasta 6 valores para que seleccione al 4 (6)
               #De 2 en dos, para que cumpla con lo solicitado (2)
```

```python
#Seleccionar todos los valores con un indice par
enteros[::2] #:: Significa que debe agarrar tooda la lista
             #2 Debe ir de dos en dos, para cumplir la condición par
```

```python
#Seleccionar todos los valores indice impar
impares = enteros[enteros % 2 !=0]
print(impares)
```
